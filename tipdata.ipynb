# **Plotting diffrent graphs to find the type of customer crowd the waiter addresses over the span of time she has collected this data**

```python
sns.barplot(x = df['Total People in the Group'], y = df.index)
```

```python
fig1=px.pie(df, values=df.index, names='Gender of the Customer', title ='Gender percentage of the customer' )
fig1.show(renderer='png')
```

# **Plotting graphs to find correlation between different variables and how they correlate with the amount tipped**

```python
plt.figure(figsize=(13,7))
sns.set_context("paper", rc={"font.size":20,"axes.titlesize":20,"axes.labelsize":18}) 
sns.regplot(x = df['Total Bill of the Customer'], y = df['The amount he was tipped by the customer'])
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot = True)
```

```python
plt.figure(figsize=(20,15))
sns.lmplot(x ='Total Bill of the Customer', y ='The amount he was tipped by the customer', hue ='Type of food item orderd', data = df)
```

```python
plt.figure(figsize=(30,15))
sns.lmplot(x ='Total Bill of the Customer', y ='The amount he was tipped by the customer', hue ='If the customer is a smoker', data = df)
```

```python
sns.lmplot(x ='Total Bill of the Customer', y ='The amount he was tipped by the customer', hue ='Gender of the Customer', data = df)
```

```python
plt.figure(figsize=(13,9))
sns.set_context("paper", rc={"font.size":15,"axes.titlesize":15,"axes.labelsize":15}) 
sns.scatterplot(x ='Total Bill of the Customer', y ='The amount he was tipped by the customer', hue ='Day of the week when he was tipped', data = df)
```
